```python
dataFrame1 = pd.read_csv("ratings.data", sep ="\t", skipinitialspace=True)
dataFrame1.columns = ['user_id', 'movie_id','rating', 'time_stamp']

dataFrame1.head()
```

```python
dataFrame2 = pd.read_csv("item.item", sep ="|", skipinitialspace=True, encoding='latin-1')
dataFrame2.columns = ['movie_id','title', 'release_date','video_release_date',
              'IMDb URL', 'unknown' , 'Action' , 'Adventure' , 'Animation' ,
              'Children', 'Comedy' , 'Crime' ,'Documentary' , 'Drama' , 'Fantasy' ,
              'Film_Noir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'Sci_Fi' ,
              'Thriller' , 'War' , 'Western' ]

dataFrame2.head()
```

```python
dataFrame3 = pd.merge(dataFrame1,dataFrame2, on='movie_id', how='inner') 
output = dataFrame3[['user_id','movie_id', 'title', 'rating','unknown' , 'Action' , 'Adventure' , 'Animation' ,
              'Children', 'Comedy' , 'Crime' ,'Documentary' , 'Drama' , 'Fantasy' ,
              'Film_Noir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'Sci_Fi' ,
              'Thriller' , 'War' , 'Western']]

output.to_csv(r'C:\Users\munat\Desktop\MapReduce\movies.csv', sep='\t', header=False, index=None)


```
